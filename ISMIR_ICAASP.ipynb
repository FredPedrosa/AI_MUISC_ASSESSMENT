```python
import fitz  # PyMuPDF
import pandas as pd
import os

# Lista dos arquivos que você baixou (confirmando nomes da imagem)
arquivos_proceedings = [
    "2021_Proceedings_ISMIR.pdf",
    "2022_Proceedings_ISMIR.pdf",
    "2023_Proceedings_ISMIR.pdf",
    "2024_Proceedings_ISMIR.pdf",
    "2025_Proceedings_ISMIR.pdf"
]

# Palavras-chave para busca combinada
# (Precisa ter um termo de GERAÇÃO + um termo de AVALIAÇÃO na mesma página)
termos_geracao = ["generation", "generative", "synthesis", "transformer", "diffusion", "composition"]
termos_humanos = ["human evaluation", "user study", "listening test", "subjective", "questionnaire", "likert", "mushra", "participants"]

def minerar_proceedings(lista_arquivos):
    resultados = []

    for nome_arquivo in lista_arquivos:
        if not os.path.exists(nome_arquivo):
            print(f"⚠️ Arquivo não encontrado: {nome_arquivo}")
            continue

        print(f"🔍 Analisando {nome_arquivo}...")
        doc = fitz.open(nome_arquivo)

        for num_pagina in range(len(doc)):
            pagina = doc.load_page(num_pagina)
            texto = pagina.get_text().lower()

            # Verifica se a página tem termos de geração
            tem_geracao = any(g in texto for g in termos_geracao)
            # Verifica se a página tem termos de avaliação humana
            tem_humano = any(h in texto for h in termos_humanos)

            if tem_geracao and tem_humano:
                # Tenta capturar o título do artigo (geralmente as primeiras linhas da página de início)
                # Como é um proceeding, os artigos começam em páginas específicas.
                snippet = texto[:200].replace('\n', ' ')

                resultados.append({
                    "Arquivo": nome_arquivo,
                    "Página": num_pagina + 1,
                    "Snippet": snippet
                })
        doc.close()

    return pd.DataFrame(resultados)

# Executar a mineração
df_minerado = minerar_proceedings(arquivos_proceedings)

# Exibir resultados
if not df_minerado.empty:
    print(f"\n✅ Mineração concluída! Encontradas {len(df_minerado)} páginas com potencial interesse.")
    # Agrupar por arquivo para ver a distribuição
    display(df_minerado.groupby('Arquivo')['Página'].count())
else:
    print("\n❌ Nenhum trecho encontrado com essa combinação de termos.")
```

```python
import fitz
import re

# Palavras-chave que indicam a pergunta do questionário
keywords_pergunta = [
    "rate the", "asked to", "scale from", "likert", "participants were asked",
    "to what extent", "please rate", "how much", "evaluate the"
]

# Atributos comuns em IHC musical
atributos = [
    "musicality", "naturalness", "creativity", "quality", "coherence",
    "fluency", "pleasantness", "human-likeness", "preference", "likability"
]

def extrair_itens_ihc(arquivos):
    banco_de_perguntas = []

    for pdf in arquivos:
        doc = fitz.open(pdf)
        for page in doc:
            text = page.get_text()
            # Divide o texto em sentenças (aproximadamente)
            sentences = re.split(r'\. |\n', text)

            for i, sent in enumerate(sentences):
                sent_low = sent.lower()

                # Se a frase contém termos de questionário + atributos musicais
                if any(k in sent_low for k in keywords_pergunta) and any(a in sent_low for a in atributos):
                    # Pega a frase atual e a próxima para garantir o contexto
                    contexto = sent + ". " + (sentences[i+1] if i+1 < len(sentences) else "")
                    banco_de_perguntas.append({
                        "Ano": pdf[:4],
                        "Item_Encontrado": contexto.strip()
                    })
        doc.close()
    return banco_de_perguntas

perguntas = extrair_itens_ihc(arquivos_proceedings)
df_perguntas = pd.DataFrame(perguntas)
df_perguntas.to_csv("itens_questionarios_ismir.csv", index=False)

# Mostrar as primeiras 20 "perguntas" encontradas
for i, row in df_perguntas.head(20).iterrows():
    print(f"[{row['Ano']}] {row['Item_Encontrado']}\n")
```

```python
import fitz
import re
import pandas as pd

# Dicionário de construtos comuns em IHC/Music Generation
constructs_keywords = {
    "Naturalness": ["naturalness", "human-like", "realism"],
    "Musicality": ["musicality", "musical quality", "harmony"],
    "Preference": ["preference", "likeability", "enjoyment"],
    "Fluency": ["fluency", "smoothness", "coherence"],
    "Creativity": ["creativity", "originality", "novelty"],
    "Fidelity": ["audio quality", "perceptual quality", "fidelity"]
}

def extrair_construtos_itens(arquivos):
    db = []

    for pdf in arquivos:
        doc = fitz.open(pdf)
        for page in doc:
            text = page.get_text()
            # Procura por padrões onde o item é definido (ex: "Preference: how much you like it")
            # O regex busca: Palavra em negrito ou início de linha + descrição
            matches = re.findall(r'([A-Z][a-z]+):\s*(.*?\.)', text)

            for const_name, item_desc in matches:
                # Valida se o construto mapeia para nossas categorias de música/IA
                const_low = const_name.lower()
                for construct, keywords in constructs_keywords.items():
                    if any(k in const_low for k in keywords) or any(k in item_desc.lower() for k in keywords):
                        db.append({
                            "Ano": pdf[:4],
                            "Construto": construct,
                            "Item_Original": item_desc.strip(),
                            "Contexto": const_name
                        })
        doc.close()
    return pd.DataFrame(db)

# Executar na sua lista de proceedings
df_ihc = extrair_construtos_itens(arquivos_proceedings)

# Limpar duplicatas e exibir
df_ihc = df_ihc.drop_duplicates(subset=['Item_Original'])
display(df_ihc.sort_values(by="Construto"))
```

```python
import fitz
import re

def extrair_github_links(arquivos):
    links_encontrados = {}

    for pdf in arquivos:
        doc = fitz.open(pdf)
        links_pdf = []
        for page in doc:
            text = page.get_text()
            # Regex para encontrar links do github
            found = re.findall(r'github\.com/[\w\-/]+', text)
            links_pdf.extend(found)

        # Remove duplicatas e salva
        links_encontrados[pdf] = list(set(links_pdf))
        doc.close()
    return links_encontrados

github_dict = extrair_github_links(arquivos_proceedings)

for pdf, links in github_dict.items():
    print(f"\n🔗 Links em {pdf}:")
    for link in links:
        print(f"   https://{link}")
```
